```python
import pandas as pd
```

```python
data = pd.read_csv('Friends.csv')
```

```python
data
```

```python
data_dict = {}
for row in data.iterrows():
    temp_list_for_each_person = []
    temp_list_for_each_person.append(str(row[1]["Happy"]))
    temp_list_for_each_person.append(str(row[1]["Sad"]))
    temp_list_for_each_person.append(str(row[1]["Angry"]))
    temp_list_for_each_person.append(str(row[1]["Neutral"]))
    temp_list_for_each_person.append(str(row[1]["Emotional"]))
    export_string = '#'.join(temp_list_for_each_person)
    data_dict[row[1]["Full Name"]] = export_string
    
data_dict
```

```python
import json 
with open("friends.txt", "w") as fp:
    json.dump(data_dict, fp)  # encode dict into JSON
print("Done writing dict into .txt file")
```
