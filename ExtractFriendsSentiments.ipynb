```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Scores.csv')
```

```python
data
```

```python
data_dict = {}
for row in data.iterrows():
    temp_list_for_each_person = []
    happy_score = row[1]["Happy Score"]
    temp_list_for_each_person.append(str(row[1]["Happy Score"]))
    temp_list_for_each_person.append(str(row[1]["Sad Score"]))
    temp_list_for_each_person.append(str(row[1]["Angry Score"]))
    temp_list_for_each_person.append(str(row[1]["Neutral Score"]))
    temp_list_for_each_person.append(str(row[1]["Passionate Score"]))
    export_string = '#'.join(temp_list_for_each_person)
    data_dict[row[1]["Full Name"]] = export_string
    
data_dict
```

```python
import json 
with open("friends.txt", "w") as fp:
    json.dump(data_dict, fp)  # encode dict into JSON
print("Done writing dict into .txt file")
```
